### Ejercicio 4: Ionosfera (nivel avanzado)
En el archivo ionosphere.csv hay datos de la ionosfera. Las 34 primeras columnas son
valores numéricos y la última es una etiqueta de texto con ‘g’ cuando el resultado es
bueno y ‘b’ cuando es malo. Importar los datos y obtener el mejor de los siguientes
modelos:
• regresión logística
• máquinas de vector soporte con kernel: linear, poly, rbf o sigmoid
• k-nn con entre 1 y 15 vecinos
• Radius Neighbors con radio 10, 15, 20 y 25.

Los datos se han de dividir en un conjunto de entrenamiento que contenga el 80% de
los datos y el resto para test.

```python
esfera = pd.read_csv('ionosphere.csv',header=None)

# transformar categórico a numérico
resultado = {34:{"g": 1, "b": 0}}
esfera.replace(resultado, inplace=True)
esfera.head(5)
```

Cuestiones:
1. ¿Cuál es el mejor modelo en cada una de las familias?
2. ¿Cual es el mejor modelo de todas las familias?

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import *


def metricas_modelos(y_true, y_predict):
    from sklearn.metrics import accuracy_score, auc, confusion_matrix, f1_score, precision_score, recall_score, roc_curve

    # Obtención de matriz de confusión
    confusion_matrix = confusion_matrix(y_true, y_predict)

    print('La matriz de confusión es')
    print(confusion_matrix)

    print('Precisión:', accuracy_score(y_true, y_predict))
    print('Exactitud:', precision_score(y_true, y_predict))
    print('Exhaustividad:', recall_score(y_true, y_predict))
    print('F1:', f1_score(y_true, y_predict))

    false_positive_rate, recall, thresholds = roc_curve(y_true, y_predict)
    roc_auc = auc(false_positive_rate, recall)

    print('AUC:', auc(false_positive_rate, recall))

    plot(false_positive_rate, recall, 'b')
    plot([0, 1], [0, 1], 'r--')
    title('AUC = %0.2f' % roc_auc)
```

```python
# Separación de la variable objetivo y las explicativas
from sklearn.model_selection import train_test_split
target = 34
features = list(esfera.columns)
features.remove(34)

x = esfera[features]
y = esfera[target]

# Obtencion del conjunto de datos para validación
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8,test_size=0.2)
```

#### Regresión Logística


```python
from sklearn.linear_model.logistic import LogisticRegression

# Con un modelo de train
model = LogisticRegression()
model.fit(x_train, y_train)
y_predict = model.predict(x_train)
scores = cross_val_score(model, x_train, y_train)

# Evaluación de R2
print('R2 en entrenamiento es: ', model.score(x_train, y_train))
print('La media de los R2 es:', scores.mean())
#metricas
metricas_modelos(y_train, y_predict)
```

```python
# Con un modelo de test
model = LogisticRegression()
model.fit(x_test, y_test)
y_predict = model.predict(x_test)
scores = cross_val_score(model, x_test, y_test)

# Evaluación de R2 de test
print('R2 en validacion es: ', model.score(x_test, y_test))
print('La media de los R2 es:', scores.mean())
# métricas de test
metricas_modelos(y_test, y_predict)
```

En regresión logística, el único modelo analizado arroja una precision de 0.94 AUC de 0.93 y una media de los R2 es: 0.815

#### Máquina de soportes para entrenamiento

```python
from sklearn.svm import SVC
values = ('linear','poly','rbf','sigmoid')

for i in values:
    svc = SVC(kernel = i)
    svc.fit(x_train, y_train)
    y_predict_train = svc.predict(x_train)
    scores = cross_val_score(svc, x_train, y_train, cv=5, scoring='roc_auc')
    print('Para kernel =', i, 'las métricas de entrenamiento son:')
    scores_train = metricas_modelos(y_train, y_predict_train)
    print('La media de los R2 es:', scores.mean())
    print('---------------------------------------------------------------')
   
    
```

#### Máquina de soportes para validación

```python
from sklearn.svm import SVC
values = ('linear','poly','rbf','sigmoid')

for i in values:
    svc = SVC(kernel = i)
    svc.fit(x_test, y_test)
    y_predict_test = svc.predict(x_test)
    scores = cross_val_score(svc, x_test, y_test, cv=5, scoring='roc_auc')
    print('Para kernel =', i, 'las métricas de validación son:')
    scores_train = metricas_modelos(y_test, y_predict_test)
    print('La media de los R2 es:', scores.mean())
    print('---------------------------------------------------------------')
   
```

#### Análisis de variables categóricas


Para las variables categóricas se puede obtener la tabla de frecuencias de cada una de ella para ver la posibilidad de que sean incluidas en el modelo y si pueden presentar problemas.

En las variables 'STUDYYEAR' podría presentar niveles que tienen pocos registros significativos, para lo cual se podría dar una agrupación de niveles con el WoE

```python
def get_WoE(data, var, target):
    crosstab = pd.crosstab(data[target], data[var])
    
    print("Obteniendo el Woe para la variable", var, ":")
    
    for col in crosstab.columns:
        if crosstab[col][1] == 0:
            print("  El WoE para", col, "[", sum(crosstab[col]), "] es infinito")
        else:
            print("  El WoE para", col, "[", sum(crosstab[col]), "] es", np.log(float(crosstab[col][0]) / float(crosstab[col][1])))
```

```python
get_WoE(miopia, 'STUDYYEAR', target)
```

Vemos que los WOE son similares para algunos años y significativos. Consideramos que no son muchos niveles, por lo que no modificaremos la variable

```python
get_WoE(miopia, 'AGE', target)
```

Vemos que el nivel 9 es bajo y solo tiene 6 datos. Por lo cual podría convenir agrupar por WoE parecidos

```python
data_miopia = pd.concat((pd.get_dummies(miopia['GENDER'], prefix = 'GENDER'),
                        pd.get_dummies(miopia['MOMMY'], prefix = 'MOMMY'),
                        pd.get_dummies(miopia['DADMY'], prefix = 'DADMY'),
                        pd.get_dummies(miopia['STUDYYEAR'], prefix = 'STUDYYEAR'),
                        pd.get_dummies(miopia['AGE_GR'], prefix = 'AGE'),
                        pd.get_dummies(miopia['SPORTHR_GRP'], prefix = 'SPORTHR'),
                        pd.get_dummies(miopia['READHR_GRP'], prefix = 'READHR'),
                        pd.get_dummies(miopia['COMPHR_GRP'], prefix = 'COMPHR'),
                        pd.get_dummies(miopia['STUDYHR_GRP'], prefix = 'STUDYHR'),
                        pd.get_dummies(miopia['TVHR_GRP'], prefix = 'TVHR'),
                        pd.get_dummies(miopia['DIOPTERHR_GRP'], prefix = 'DIOPTERHR'),
                        pd.get_dummies(miopia['SPHEQ_2'], prefix = 'SPHEQ'),
                        pd.get_dummies(miopia['ACD_2'], prefix = 'ACD'),
                        pd.get_dummies(miopia['LT_2'], prefix = 'LT'),
                        pd.get_dummies(miopia['VCD_2'], prefix = 'VCD')), axis = 1)
```

```python
from sklearn.linear_model import LinearRegression

def calculateVIF(data):
    features = list(data.columns)
    num_features = len(features)
    
    model = LinearRegression()
    
    result = pd.DataFrame(index = ['VIF'], columns = features)
    result = result.fillna(0)
    
    for ite in range(num_features):
        x_features = features[:]
        y_featue = features[ite]
        x_features.remove(y_featue)
        
        x = data[x_features]
        y = data[y_featue]
        
        model.fit(data[x_features], data[y_featue])
        
        result[y_featue] = 1/(1 - model.score(data[x_features], data[y_featue]))
    
    return result

def selectDataUsingVIF(data, max_VIF = 5):
    result = data.copy(deep = True)
    
    VIF = calculateVIF(result)
    
    while VIF.as_matrix().max() > max_VIF:
        col_max = np.where(VIF == VIF.as_matrix().max())[1][0]
        features = list(result.columns)
        features.remove(features[col_max])
        result = result[features]
        
        VIF = calculateVIF(result)
        
    return result
```

Obtener la precisión, exactitud y exhaustividad en entrenamiento y test. ¿Por
qué se producen estos resultados?

```python
%pylab
%matplotlib inline

def metricas_modelos(y_true, y_predict):
    from sklearn.metrics import accuracy_score, auc, confusion_matrix, f1_score, precision_score, recall_score, roc_curve

    # Obtención de matriz de confusión
    confusion_matrix = confusion_matrix(y_true, y_predict)

    print('La matriz de confusión es')
    print(confusion_matrix)

    print('Precisión:', accuracy_score(y_true, y_predict))
    print('Exactitud:', precision_score(y_true, y_predict))
    print('Exhaustividad:', recall_score(y_true, y_predict))
    print('F1:', f1_score(y_true, y_predict))

    false_positive_rate, recall, thresholds = roc_curve(y_true, y_predict)
    roc_auc = auc(false_positive_rate, recall)

    print('AUC:', auc(false_positive_rate, recall))

    plot(false_positive_rate, recall, 'b')
    plot([0, 1], [0, 1], 'r--')
    title('AUC = %0.2f' % roc_auc)
```

```python
from sklearn.linear_model.logistic import LogisticRegression

model = LogisticRegression().fit(x_train, y_train)
y_pred_train = model.predict(x_train)
y_pred_test = model.predict(x_test)

metricas_modelos(y_train, y_pred_train)
```

```python
metricas_modelos(y_test, y_pred_test)
```
